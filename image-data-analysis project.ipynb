# Task 1: Data Import and Cleaning


```python
import numpy as np
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import seaborn as sns

# Load the image using PIL
path="/Users/bashayeralemadi/Desktop/PETE-219(Seers)/Group 4/berea8bit.tif"
image_path = path 
original_image = Image.open(image_path)

# Convert the image to grayscale
gray_image = original_image.convert('L')

# Denoising using a median filter
denoised_image = gray_image.filter(ImageFilter.MedianFilter(size=3))


```

# Task 2: Data Visualization


```python
# Basic visualization of grayscale and binary images
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(np.asarray(original_image), cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(np.asarray(gray_image), cmap='gray')
plt.title('Grayscale Image')

# Binarize the image
binary_threshold = 128
binary_image = gray_image.point(lambda p: p > binary_threshold and 255)
plt.subplot(1, 3, 3)
plt.imshow(np.asarray(binary_image), cmap='gray')
plt.title('Binary Image')
# Generate random pore sizes for demonstration
np.random.seed(0)
pore_sizes = np.random.randint(50, 500, 100)  # Replace this with your actual pore size data

sns.histplot(pore_sizes, bins=30, kde=True)
plt.title('Pore Size Distribution')
plt.xlabel('Pore Size')
plt.ylabel('Frequency')
plt.show()
```

# Task 3: Data Analysis


```python
# Simple porosity estimate
total_volume = np.sum(np.asarray(binary_image) == 0)
solid_volume = np.sum(np.asarray(binary_image) == 255)
porosity = ((total_volume - solid_volume) / total_volume) * 100

print(f'Simple Porosity Estimate: {porosity:.2f}%')
```
